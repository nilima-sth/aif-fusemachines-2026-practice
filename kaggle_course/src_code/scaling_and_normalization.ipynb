```python
import numpy as np
import pandas as pd
```

```python
from scipy import stats #for box-cox transformation: used to make data more “normal” (bell-shaped)
```

```python
# for min_max scaling
from mlxtend.preprocessing import minmax_scaling
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

### Scaling

```python
#Creating random exponential data of size 1000

og_data=np.random.exponential(size=1000)

#min_max scaling 
scaled_data=minmax_scaling(og_data,columns=[0])  #scaled in the range of 0->1
```

```python
#visualizing and plotting
fig, ax = plt.subplots(1,2,figsize=(15,3))

#Plotting the og_data (Original Data)
sns.histplot(og_data,ax=ax[0],legend=False,kde=True)
ax[0].set_title("Original Data")

#Plotting the scaled data 
sns.histplot(scaled_data,ax=ax[1],legend=False,kde=True)
ax[1].set_title("Scaled Data")


plt.show()
```

Here, the range was 0 to 6 and changed to 0 to 1. 

### Normalization

```python
#box cox transformation
normalized_data=stats.boxcox(og_data)
```

```python
fig, ax=plt.subplots(1,2, figsize=(15,3))

#Visualizing Original Data 
sns.histplot(og_data,ax=ax[0],kde=True,legend=False)
ax[0].set_title("Original Data")

#visualizing normalized data
sns.histplot(normalized_data,ax=ax[1],kde=True,legend=False)
ax[1].set_title("Normalized Data")

plt.show()
```
